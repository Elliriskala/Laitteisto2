```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Kpl 5.3, teht. 8 s. 96: Piirr채 funktion kuvaaja.

x = np.linspace(-1, 1)
y = np.where((-1 < x) & (x < 1), 4, np.nan)

x2 = np.linspace(1, 3)
y2 = -2*x2+6

plt.plot(x, y, color='blue')
plt.plot(x2, y2, color='blue')

plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Samoin teht채v채ss채 9

x = np.linspace(-5, 0)
y = (3/5*x)+3

x2 = np.linspace(0, 4)
y2 = np.full_like(x2, 3)

x3 = np.linspace(4, 8)
y3 = (3/4)*(8 - x3)

x4 = np.linspace(8, 10)
y4 = (1/2*x4)-4

plt.plot(x, y, color='blue')
plt.plot(x2, y2, color='blue')
plt.plot(x3, y3, color='blue')
plt.plot(x4, y4, color='blue')

plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Kpl 5.4 s 100: Teht채v채st채 1 taitanee riitt채채 kuvaajat kohdista a ja b

# a)
x = np.linspace(-2, 6)
y = x**2 - 4*x + 3

x_h = 2
y_h = x_h**2 - 4*x_h + 3
plt.scatter(x_h, y_h, color='red')

plt.plot(x, y)
plt.show()
```
